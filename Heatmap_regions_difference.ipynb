```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python


sns.set(style="dark")
regions = pd.read_csv('heatmap_regions_difference.csv', index_col='Lineage')

g = sns.relplot(x="Region", y="Lineage",
                hue="Percentage", size="Percentage",
                hue_norm=(0, 150), size_norm=(0, 100),
                palette="rocket", sizes=(0, 100),
                marker="s", linewidth=0, legend=False,
                aspect=1.5, data=regions)

ghost = g.ax.scatter([], [], c=[], vmin=0, vmax=150, cmap="rocket")
g.fig.colorbar(ghost)
# Adjust ticks to center scatterplot markers
g.ax.set_xticks([t + 0.5 for t in g.ax.get_xticks()], minor=True)
g.ax.set_yticks([t + 0.5 for t in g.ax.get_yticks()], minor=True)

# Set the limits for the x-axis and y-axis, to center and adjust the squares withint the grids
x_max = 20
y_max = 29
g.ax.set_xlim([-0.5, x_max - 0.5])
g.ax.set_ylim([-0.5, y_max - 0.5])

# Turn off major gridlines and turn on minor gridlines
g.ax.grid(False, 'major')
g.ax.grid(True, 'minor')

plt.xticks(rotation=45,horizontalalignment='right')

# Invert y axis if needed
g.ax.invert_yaxis()
# Set the plot size in centimeters
fig = g.fig
fig.set_size_inches(8, 7)  # Adjust the dimensions as desired

# Save the plot as a PDF
plt.savefig('heatmap_regions_difference.pdf',dpi=300)

# Display the plot
plt.show()
```
